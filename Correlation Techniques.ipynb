```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('mtcars.csv')
```

```python
df
```

```python
x= df.drop('mpg',axis =1)
```

```python
correlation_matrix=x.corr()
correlated_featurs=set()
```

```python
%precision 2
for i in range(len(correlation_matrix)):
    for j in range(i):
        if  abs(correlation_matrix.iloc[i,j])>0.8:
            correlated_featurs.add(correlation_matrix.columns[i])
```

```python
df.drop(correlated_featurs, axis=1, inplace=True)
```

```python
df
```

```python
df.corr()
```

```python
from scipy.stats import spearmanr,pearsonr
```

```python
df= pd.DataFrame({
    'X' : x,
    'Y' : y,
    'Z' : z,
    'P' : p
})
```

```python
spearmanr(x,y)
```

```python
pearsonr(rankdata(x),rankdata(y))
```

```python
df.corr(method='spearman')
```
