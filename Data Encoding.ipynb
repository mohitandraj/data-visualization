#### Label Encoding

```python
df = pd.read_csv('student.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Create the Object
le = LabelEncoder()
```

```python
df['name']=le.fit_transform(df['name'])
```

```python
df
```

```python
new = 'anil'
```

```python
le.transform([new])
```

```python
le.inverse_transform([5])
```
